```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
filename = "auto.csv"
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
#reading csv and creating dataframe
df = pd.read_csv(filename,header=None)
```

```python
df.columns = headers
```

```python
# replace "?" to NaN
df.replace("?", np.nan, inplace = True)
df.head(5)
```

```python
#evaluating missing data
missing_data = df.isnull()
missing_data.head(10)
```

```python
missing_data_list = missing_data.columns.values.tolist()
missing_data_list
```

```python
for column in missing_data_list:
    print(column)
    print(missing_data[column].value_counts())
    
```

```python
#average of 'normalized-losses' column
avg_norm_loss = df["normalized-losses"].astype("float").mean(axis=0)
print("Average of normalized-losses:", avg_norm_loss)
```

```python
# average of 'bore' column
avg_bore=df['bore'].astype('float').mean(axis=0)
print("Average of bore:", avg_bore)
```

```python
#replace nan with mean value in 'bore' column
df.replace(np.nan,avg_bore,inplace=True)
```

```python
df['bore']
```

```python
df['stroke']
```

```python
#average of stroke column
avg_stroke = df['stroke'].astype(float).mean(axis=0)
avg_stroke
```

```python
#replace nan values with mean
df['stroke'].replace(np.nan,avg_stroke)
```

```python
avg_horsepower = df['horsepower'].astype('float').mean(axis=0)
print("Average horsepower:", avg_horsepower)
```

```python
df['horsepower'].replace(np.nan, avg_horsepower, inplace=True)
```

```python
avg_peakrpm=df['peak-rpm'].astype('float').mean(axis=0)
print("Average peak rpm:", avg_peakrpm)
```

```python
df['peak-rpm'].replace(np.nan, avg_peakrpm, inplace=True)
```

```python
#count the values in num-of-doors columns
df['num-of-doors'].value_counts()
```

```python
#most common type or object with maximum values
df['num-of-doors'].value_counts().idxmax()
```

```python
df["num-of-doors"].replace(np.nan, "four", inplace=True)
```

```python
# simply drop whole row with NaN in "price" column
df.dropna(subset=["price"], axis=0, inplace=True)

# reset index, because we droped two rows
df.reset_index(drop=True, inplace=True)
```

```python
#to check the datatypes
df.dtypes
```

```python
# Convert mpg to L/100km by mathematical operation (235 divided by mpg)
df['city-L/100km'] = 235/df["city-mpg"]

# check your transformed data 
df.head()
```

```python
df.head()
```

```python
#normalizing
# replace (original value) by (original value)/(maximum value)
df['length'] = df['length']/df['length'].max()
df['width'] = df['width']/df['width'].max()
```

```python
df['height']
```

```python
#normalizing height
df['height']=df['height']/df['height'].max()
```

```python
df['height']
```

```python
#create bin array
df['horsepower'] = df['horsepower'].astype(float)
bins = np.linspace(min(df["horsepower"]), max(df["horsepower"]), 4)
bins
```

```python
#apply the function "cut" to determine what each value of df['horsepower'] belongs to
group_names = ['Low', 'Medium', 'High']
df['horsepower-binned'] = pd.cut(df['horsepower'], bins, labels=group_names, include_lowest=True )
df[['horsepower','horsepower-binned']].head(20)
```

```python
df["horsepower-binned"].value_counts()
```

```python
plt.bar(group_names, df["horsepower-binned"].value_counts())

# set x/y labels and plot title
plt.xlabel("horsepower")
plt.ylabel("count")
plt.title("horsepower bins")
```

```python
# draw historgram of attribute "horsepower" with bins = 3
plt.hist(df["horsepower"], bins = 5)

# set x/y labels and plot title
plt.xlabel("horsepower")
plt.ylabel("count")
plt.title("horsepower bins")
```
